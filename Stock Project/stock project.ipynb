#### The objective is to ascertain the highest Closing price for the stocks of each bank during the specified time frame.
</span>

```python
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

<span style="font-family:Times New Roman">  
    
#### The equation below displays the profit gained (r) during a particular timeframe. The formula involves using "p" to denote the price, which is then further specified as time "t." In this scenario, the focus is on dates, indicating the price on a particular date divided by the price on the preceding date, with one subtracted from the result. In essence, it figures out the rate of return in percentage form. 

#### $$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

#### The first step involves creating a dataframe that is empty.  The subsequent move is to apply the pct_change() method from pandas to the Close column, resulting in a new column that shows the return value. Following this, we should construct a for loop that will cycle through every Bank Stock Ticker and create a returns column, which will then be appended as a fresh column in the returns DataFrame.


</span>

```python
returns = pd.DataFrame()
for tick in tickers:
    returns[tick +r' '+' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

## Moving Averages  
#### It is important to analyze the mean values of stock fluctuations in the year 2008. Our aim is to compare the 30-day moving average with the closing price of Bank Of America's stock throughout 2008. We employ the rolling method to accomplish this.

</span>

```python
BAC = BAC.sort_index()
plt.figure(figsize=(12,4))
BAC['Close'].loc['2008-01-01':'2009-01-01'].rolling(window=30).mean().plot(label = '30 day Moving Avrage')
BAC['Close'].loc['2008-01-01':'2009-01-01'].plot(label='BAC Close')
plt.legend()
```

####  In the forthcoming document, our aim is to construct a heatmap using Plotly.



```python
import plotly.graph_objs as go
from plotly.offline import iplot

close_corr = bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr()
heatmap = go.Heatmap(z=close_corr.values,
    x=close_corr.columns,
    y=close_corr.index,
    colorscale='RdYlBu'
)

layout = go.Layout(width=600, height=600 )
# Create a figure
fig = {
    'data': [heatmap],
    'layout': layout
}


# Show the plot using iplot
iplot(fig)
```
